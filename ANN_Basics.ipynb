```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
```

```python
data = fetch_california_housing()
```

```python
X_train_full, X_test, y_train_full, y_test = train_test_split(data.data, data.target,random_state=42)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train_full, y_train_full,random_state=42)
```

```python
Scale = StandardScaler()
```

```python
X_train = Scale.fit_transform(X_train)
X_valid = Scale.transform(X_valid)
X_test = Scale.transform(X_test)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(30, activation="relu", input_shape=X_train.shape[1:]),
    tf.keras.layers.Dense(1)
])
model.compile(loss="mean_squared_error", optimizer="sgd")
history = model.fit(X_train, y_train, epochs=20, validation_data=(X_valid, y_valid))
mse_test = model.evaluate(X_test, y_test)
```

```python
print(mse_test)
```

```python
print('Making predictions on the test set:')
y_pred = model.predict(X_test)
print(y_pred[:5]) # Print the first 5 predictions

print('\nModel Summary:')
model.summary()
```

## Create a Gradio Interface for the Model

```python
import gradio as gr

def predict_housing_price(
    MedInc,
    HouseAge,
    AveRooms,
    AveBedrms,
    Population,
    AveOccup,
    Latitude,
    Longitude
):
    # Create a numpy array from the input features
    features = np.array([
        MedInc,
        HouseAge,
        AveRooms,
        AveBedrms,
        Population,
        AveOccup,
        Latitude,
        Longitude
    ]).reshape(1, -1)

    # Scale the input features using the pre-fitted StandardScaler
    scaled_features = Scale.transform(features)

    # Make a prediction using the trained model
    prediction = model.predict(scaled_features)[0][0]

    return float(prediction)

# Create Gradio input components for each feature
inputs = [
    gr.Number(label='Median Income (MedInc)'),
    gr.Number(label='Median House Age (HouseAge)'),
    gr.Number(label='Average Rooms (AveRooms)'),
    gr.Number(label='Average Bedrooms (AveBedrms)'),
    gr.Number(label='Population (Population)'),
    gr.Number(label='Average Occupancy (AveOccup)'),
    gr.Number(label='Latitude (Latitude)'),
    gr.Number(label='Longitude (Longitude)')
]

# Create Gradio output component
output = gr.Number(label='Predicted Housing Price (in 100,000s USD)')

# Create and launch the Gradio interface
interface = gr.Interface(
    fn=predict_housing_price,
    inputs=inputs,
    outputs=output,
    title='California Housing Price Predictor',
    description='Enter the housing features to predict the median house value.'
)

interface.launch()
```

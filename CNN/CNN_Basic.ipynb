```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout,MaxPooling2D,Conv2D,Flatten
from tensorflow.keras.utils import to_categorical
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("oddrationale/mnist-in-csv")

print("Path to dataset files:", path)

```

```python
import os

df = pd.read_csv(os.path.join(path, "mnist_train.csv"))
df_test = pd.read_csv(os.path.join(path, "mnist_test.csv"))
```

```python
df.head()
```

```python
x_tarin = df.drop('label',axis=1)
y_train = df['label'].values
x_test = df_test.drop('label',axis=1)
y_test = df_test['label'].values
```

```python
x_tarin = x_tarin.astype('Float32')/255
x_test = x_test.astype('Float32')/255
```

```python
x_train_img = x_tarin.values.astype(np.float32).reshape(-1,28,28,1)
x_test_img = x_test.values.astype(np.float32).reshape(-1,28,28,1)
```

```python
y_train_original_labels = df['label'].values
y_test_original_labels = df_test['label'].values

y_train = to_categorical(y_train_original_labels, 10)
y_test = to_categorical(y_test_original_labels, 10)
```

```python
CNN = Sequential()
CNN.add(Conv2D(filters=32,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'))
CNN.add(MaxPooling2D(pool_size=(2,2)))
CNN.add(Conv2D(filters=64,kernel_size=(3,3),input_shape=(28,28,1),activation='relu'))
CNN.add(MaxPooling2D(pool_size=(2,2)))
CNN.add(Flatten())
CNN.add(Dense(128,activation='relu'))
CNN.add(Dense(10,activation='softmax'))
CNN.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
History = CNN.fit(x_train_img,y_train,epochs=10,batch_size=100,validation_data=(x_test_img,y_test))
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Dense
```

```python
arr = np.array([[1,2,3,4],[2,4,6,8],[3,6,9,12],[5,10,15,20]])
next_arr = np.array([5,10,25,15])
```

```python
model = Sequential()
model.add(SimpleRNN(units=10,input_shape=(arr.shape[1],1),activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
arr = arr.reshape((-1,arr.shape[1],1))
```

```python
history = model.fit(arr,next_arr,epochs=100,verbose=1)
```

```python
new_arr = np.array([[4,6,8,10]])
new_arr = new_arr.reshape((-1,new_arr.shape[1],1))
```

```python
predicted_num=model.predict(new_arr)
```

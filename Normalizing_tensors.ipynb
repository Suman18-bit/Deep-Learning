```python
import tensorflow as tf
```

```python
tensor = tf.random.uniform(shape=(4,4),minval=0,maxval=10,dtype=tf.int32)
```

```python
print("The Tensor \n",tensor.numpy())
```

```python
normalized_tensor = tf.cast(tensor, tf.float32) / 8
```

```python
print("The Normalized Tensor \n",normalized_tensor.numpy())
```

```python
print("The MEAN of the Normalized Tensor is :-",tf.reduce_mean(normalized_tensor).numpy())
```

```python
print("The Standerd Deviation of the Normalized Tensor is :-",tf.math.reduce_std(normalized_tensor).numpy())
```

```python
print("The Variance of the Normalized Tensor is :-",tf.math.reduce_variance(normalized_tensor).numpy())
```
